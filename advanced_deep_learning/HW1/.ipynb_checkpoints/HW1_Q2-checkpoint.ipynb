```python
import matplotlib.pyplot as plt
import numpy as np
import random
from PIL import Image
import PIL.ImageOps    
import torchvision
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
import torchvision.utils
import torch
from torch.autograd import Variable
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
from tqdm.notebook import tqdm, trange
import torchvision.models as models
```

```python
ts = transforms.Compose([transforms.Resize((224,224)),transforms.ToTensor()])
train_set = datasets.StanfordCars("./dataset_q2/","train",download = False,
                                  transform=ts,
                                  target_transform = lambda x: torch.tensor(x))
val_set = datasets.StanfordCars("./dataset_q2/","test",download = False,
                                transform=ts,
                                target_transform = lambda x: torch.tensor(x))
train_loader = DataLoader(train_set,batch_size= 256,shuffle = True)
val_loader = DataLoader(val_set,batch_size=256,shuffle = True)
```

```python
model = models.efficientnet_b0(pretrained=True)
for params in model.parameters():
    params.requires_grad = False
    
model.classifier[1] = nn.Linear(in_features=1280, out_features=196)
model = model.cuda()
model
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(),lr = 1e-3,weight_decay=1e-10)

accuarcy_score = lambda x,y: torch.sum((torch.argmax(x,1)==y))/x.shape[0]
logs = {"training loss":[],"validation loss": [], "training metric": [],"validation metric":[]}
```

```python
epochs = 30
loop = trange(epochs)
for epoch in loop:
    train_running_loss = 0
    train_running_mertic = 0
    model.train()
    for X,y in train_loader:
        X = X.cuda()
        y = y.cuda()
        out = model(X)
        loss = loss_fn(out,y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        with torch.no_grad():
            train_running_loss += loss.item()
            acc = accuarcy_score(out,y)
            train_running_mertic += acc.item()
    
    train_running_loss = train_running_loss/len(train_loader)
    train_running_mertic = train_running_mertic/len(train_loader)
    val_running_loss = 0
    val_running_metric = 0
    model.eval()
    for X,y in val_loader:
        with torch.no_grad():
            X = X.cuda()
            y = y.cuda()
            out = model(X)
            loss = loss_fn(out,y)
            acc = accuarcy_score(out,y)
            val_running_loss += loss.item()
            val_running_metric += acc.item()
    val_running_loss = val_running_loss/len(val_loader)
    val_running_metric = val_running_metric/len(val_loader)
    logs["training loss"].append(train_running_loss)
    logs["training metric"].append(train_running_mertic)
    logs["validation loss"].append(val_running_loss)
    logs["validation metric"].append(val_running_metric)
    loop.set_postfix({"training loss":train_running_loss,
                     "validation loss":val_running_loss,
                     "training metric":train_running_mertic,
                     "validation metric":val_running_metric})
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2,1,1)
plt.plot(logs["training loss"],label = "training")
plt.plot(logs["validation loss"], label = "validation")
plt.title("CCE loss")
plt.legend()
plt.grid("on")
plt.subplot(2,1,2)
plt.plot(logs["training metric"])
plt.plot(logs["validation metric"])
plt.title("accuracy score")
plt.grid("on")

plt.show()
```
